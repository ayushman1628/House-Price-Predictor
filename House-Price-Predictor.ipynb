```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('/content/drive/MyDrive/Projects/House_price/train.csv')
df
```

```python
df.shape
```

```python
df.info()
```

```python
numeric_col = df.select_dtypes(include=['int64','float64']).columns
numeric_col
```

```python
df[numeric_col]= df[numeric_col].fillna(df[numeric_col].mean())
```

```python
categorical_col= df.select_dtypes(include=['object']).columns
categorical_col
```

```python
corr= df[numeric_col].corr()
corr
```

```python
df[categorical_col]= df[categorical_col].fillna('NA')
```

```python
df[categorical_col].nunique()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder= OneHotEncoder(handle_unknown='ignore')
```

```python
encoder.fit(df[categorical_col])
```

```python
encoded_col= list(encoder.get_feature_names_out(categorical_col))
print(encoded_col)
```

```python
df[encoded_col]= encoder.transform(df[categorical_col]).toarray()
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
val_df, train_df= train_test_split(df, test_size=0.8, random_state=42)
```

```python
train_df
```

```python
input_col= list(numeric_col.drop('SalePrice')) + encoded_col
input_col
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler= MinMaxScaler().fit(train_df[numeric_col.drop('SalePrice')])
```

```python
train_df[numeric_col.drop('SalePrice')]= scaler.transform(train_df[numeric_col.drop('SalePrice')])
val_df[numeric_col.drop('SalePrice')]= scaler.transform(val_df[numeric_col.drop('SalePrice')])
```

```python
def return_mean(inputs):
    return np.full(len(inputs), train_df.SalePrice.mean())
```

```python
mean_preds = return_mean(train_df[input_col])
```

```python
from sklearn.metrics import mean_absolute_error
target_col = 'SalePrice'
```

```python
mean_absolute_error(mean_preds, train_df[target_col])
```

Linear Model


```python
from sklearn.linear_model import LinearRegression
```

```python
lin_model= LinearRegression()
```

```python
linear_model= lin_model.fit(train_df[input_col], train_df[target_col])
```

```python
linreg_train_pred= lin_model.predict(train_df[input_col])
linreg_val_pred= lin_model.predict(val_df[input_col])
```

```python
mean_absolute_error(linreg_train_pred, train_df[target_col])
```

```python
mean_absolute_error(linreg_val_pred, val_df[target_col])
```
